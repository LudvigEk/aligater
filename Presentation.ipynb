```python
import aligater as ag
%matplotlib inline
```

```python
fileList=ag.collectFiles("/media/ludvig/Project_Storage/GAIM/data/GAIM_fcs/")
```

```python
sample=ag.loadFCS("/media/ludvig/Project_Storage/GAIM/data/GAIM_fcs/01-03-2016/plate 1/B cells_A5_A05_031.fcs",return_type='agsample')
```

```python
sample()
```

```python
ag.plotHeatmap(sample(),"FSC-A","FSC-H")
```

```python
no_clutter=ag.gateThreshold(sample(),"FSC-A","FSC-H",population='lower',orientation='vertical',thresh=260000)
```

```python
PCgate=ag.gatePC(sample(),"FSC-A","FSC-H",vI=no_clutter,widthScale=3,heightScale=4)
```

```python
singlets=ag.AGgate(PCgate,sample.full_index())
```

```python
sample.update(singlets,name='singlets',parentName='ungated')
```

```python
sample.report()
```

```python
sample.printStats(output="output/testOutput.txt",header=True)
```
